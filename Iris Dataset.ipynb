### (6) Train-Test Split:

```python
X=df.drop(columns=['lables'])
Y=df['lables']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.4,random_state=7)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

### (7) Model Training:

#### (1) KNN Classifier:

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

##### Evalution and Accuracy:

```python
y_predict=model.predict(x_test)
y_predict
```

```python
from sklearn import metrics
print("Accuracy=",metrics.accuracy_score(y_test,y_predict)*100,"%")
```

#### Repeating KNN for k=3:

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
print("Accuracy=",metrics.accuracy_score(y_test,y_predict)*100,"%")
```

* We can choose KNN model with k=3 because it gives 95% testing accuracy.

##### Using Cross-validation:

```python
from sklearn.model_selection import cross_val_score
model = KNeighborsClassifier(n_neighbors=2)
scores = cross_val_score(model, X, Y, cv=10, scoring="accuracy")
scores
```

```python
print (scores.mean()*100,"%")
```

* We can got better accuracy than previous one as we got through train-test split method. 

#### (2) Decision Tree Classifier:

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(x_train,y_train)
```

##### Evalution and Accuracy:

```python
y_pred=tree.predict(x_test)
```

```python
from sklearn import metrics
print("Accuracy=",metrics.accuracy_score(y_test,y_pred)*100,"%")
```

##### Using Cross-validation:

```python
from sklearn.model_selection import cross_val_score
tree = DecisionTreeClassifier()
scores = cross_val_score(tree, X, Y, cv=10, scoring="accuracy")
scores
```

```python
print(scores.mean()*100,"%")
```

#### (3) SVM:

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(x_train,y_train)
```

```python
y_predictions = clf.predict(x_test)
```

#### Using Cross-validation:

```python
from sklearn.model_selection import cross_val_score
clf = SVC()
scores = cross_val_score(clf, X, Y, cv=10, scoring="accuracy")
scores
```

```python
print(scores.mean()*100,"%")
```
